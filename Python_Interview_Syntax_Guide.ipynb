## 1. Variables & Data Types
Python is dynamically typed.
- int
- float
- str
- bool
- NoneType

```python
x = 10          # int
y = 10.5        # float
name = "John"  # string
is_valid = True # boolean
z = None        # NoneType
```

## 2. Type Casting
Convert between types using built-in functions.

```python
a = int("5")
b = float("5.5")
c = str(100)
d = bool(1)
```

## 3. Operators
- Arithmetic: +, -, *, /, //, %, **
- Comparison: ==, !=, >, <, >=, <=
- Logical: and, or, not
- Identity: is, is not
- Membership: in, not in

```python
a = 10
b = 3
print(a // b)  # Floor division
print(a ** b)  # Power
print(a > b and b < 5)
```

## 4. Conditional Statements

```python
x = 10
if x > 5:
    print("Greater")
elif x == 5:
    print("Equal")
else:
    print("Smaller")
```

## 5. Loops
### For Loop
Iterates over iterable objects.

```python
for i in range(5):
    print(i)
```

### While Loop
Repeats while condition is True.

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

## 6. Data Structures
### List (Mutable, Ordered)

```python
lst = [1, 2, 3]
lst.append(4)
lst[0] = 10
```

### Tuple (Immutable, Ordered)

```python
tup = (1, 2, 3)
# tup[0] = 10  # Error
```

### Set (Unordered, Unique)

```python
s = {1, 2, 2, 3}
s.add(4)
```

### Dictionary (Key-Value Pairs)

```python
d = {"name": "John", "age": 30}
print(d["name"])
d["age"] = 31
```

## 7. Functions
Defined using def keyword.

```python
def greet(name):
    return f"Hello {name}"

print(greet("Alice"))
```

### Lambda Function (Anonymous Function)

```python
square = lambda x: x * x
print(square(5))
```

## 8. OOP (Object-Oriented Programming)
### Class & Object

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hi, I am {self.name}"

p = Person("John")
print(p.greet())
```

## 9. Exception Handling

```python
try:
    x = 1 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Done")
```

## 10. List Comprehension

```python
squares = [x*x for x in range(5)]
```

## 11. Generators
Use yield instead of return.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)
```
