```python

import torch
import torch.nn as nn
import torch.optim as optim
print("PyTorch version:", torch.__version__)
```


## 1. Tensors (Core Data Structure)

`torch.Tensor` is similar to NumPy arrays but supports **GPU acceleration** and **automatic differentiation**.


```python

# Creating tensors
a = torch.tensor([1, 2, 3])
b = torch.zeros((2, 3))
c = torch.ones((3, 3))
d = torch.rand((2, 2))

print(a)
print(b)
print(c)
print(d)
```


### Tensor Operations


```python

x = torch.tensor([1.0, 2.0, 3.0])
y = torch.tensor([4.0, 5.0, 6.0])

print(x + y)        # Element-wise addition
print(x * y)        # Element-wise multiplication
print(torch.dot(x, y))  # Dot product
```


## 2. Autograd (Automatic Differentiation)

PyTorch automatically computes gradients if `requires_grad=True`.


```python

x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3*x + 1
y.backward()

print("dy/dx:", x.grad)
```


## 3. Neural Networks with `nn.Module`


```python

class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(1, 10)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(10, 1)

    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        return self.linear2(x)

model = SimpleNN()
print(model)
```


## 4. Training Loop (Core PyTorch Skill)


```python

# Dummy data
X = torch.randn(100, 1)
y = 3*X + 2

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(100):
    optimizer.zero_grad()
    preds = model(X)
    loss = criterion(preds, y)
    loss.backward()
    optimizer.step()

print("Final loss:", loss.item())
```


## 5. GPU Usage


```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print("Using device:", device)
```


## 6. Custom Dataset & DataLoader


```python

from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):
    def __init__(self):
        self.x = torch.randn(100, 1)
        self.y = 2*self.x + 1

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

dataset = MyDataset()
loader = DataLoader(dataset, batch_size=16, shuffle=True)

for batch_x, batch_y in loader:
    print(batch_x.shape)
    break
```


## 7. CNN Example (Image Data)


```python

class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv2d(1, 16, 3)
        self.fc = nn.Linear(16*26*26, 10)

    def forward(self, x):
        x = self.conv(x)
        x = torch.relu(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

cnn = SimpleCNN()
print(cnn)
```


## 8. Saving & Loading Models


```python

torch.save(model.state_dict(), "model.pth")

loaded_model = SimpleNN()
loaded_model.load_state_dict(torch.load("model.pth"))
loaded_model.eval()
```
